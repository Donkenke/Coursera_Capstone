```python
latitude=43.65
longitude=-79.37
```

From now on, i will process and cluster in a very simple way. I will just order the data points according to their latitude, and then, i will pick the first 9 and cluster them in a particular category. This is an interative process till the last data point is clustered. This will result in a geographical kind of very simple method trying to create 4 equaly size clusters. 

```python
df41=df4.sort_values('Latitude')
```

```python
df3=df41.reset_index()
```

```python
df2=df3.drop(columns=['index'])
```

```python
df33=df2[0:9]
```

```python
df34=df2[10:19]
```

```python
df35=df2[20:29]
```

```python
df36=df2[30:38]
```

```python
#The clustering is mostly geographical due to the data available for this project.  

venues_map = folium.Map(location=[latitude, longitude], zoom_start=11) 


folium.features.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='Conrad Hotel',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)

for lat, lng, label in zip(df33.Latitude, df33.Longitude, df33.Borough):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)

for lat, lng, label in zip(df34.Latitude, df34.Longitude, df34.Borough):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='green',
        popup=label,
        fill = True,
        fill_color='green',
        fill_opacity=0.6
    ).add_to(venues_map)
    
for lat, lng, label in zip(df35.Latitude, df35.Longitude, df35.Borough):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='yellow',
        popup=label,
        fill = True,
        fill_color='yellow',
        fill_opacity=0.6
    ).add_to(venues_map)

for lat, lng, label in zip(df36.Latitude, df36.Longitude, df36.Borough):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='brown',
        popup=label,
        fill = True,
        fill_color='brown',
        fill_opacity=0.6
    ).add_to(venues_map)    
      
    
# display map
venues_map
```
