Try to group them in order to obrain a single Postal Code for each Borough

```python
def f(x):
     return pd.Series(dict(Borough = ", ".join(x['Borough']), 
                        Neighborhood = "%s" % ', '.join(x['Neighborhood'])))
```

```python
N9=N5.groupby('Postal Code').apply(f)
```
